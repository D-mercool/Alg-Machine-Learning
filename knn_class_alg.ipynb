### Метод k ближайших соседей

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold #Кросс-валидация
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import scale
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('wine.csv',  sep=',')
X = data.iloc[:, 1:] #Признаки
y = data['A'] #Объекты

#Разбиение на 5 групп
kf = KFold(n_splits = 5, random_state=42, shuffle=True)

#Обучение модели
mass = []
for k in range(1,51):
    kn = KNeighborsClassifier(n_neighbors=k)
    kn.fit(X,y)
    #Проверка качетсва разбиения модели на наборах
    scores = cross_val_score(kn, X, y, cv=kf)
    mass.append(scores)
```

```python
d1 = pd.DataFrame(mass, columns=['1', '2', '3', '4', '5'], index=range(1,51))
display(d1.sort_values(by=['1']))
print(d1.mean(axis=1))
```

Производим масштабирование признаков с помощью функции sklearn.preprocessing.scale. Снова найдите оптимальное k на кросс-валидации.

```python
X = scale(X)

#Обучение модели
mass = []
for k in range(1,51):
    kn = KNeighborsClassifier(n_neighbors=k)
    kn.fit(X,y)
    #Проверка качетсва разбиения модели на наборах
    scores = cross_val_score(kn, X, y, cv=kf)
    mass.append(scores)

d2 = pd.DataFrame(mass, columns=['1', '2', '3', '4', '5'], index=range(1,51))
display(d2.sort_values(by=['1']))
print(d2.mean(axis=1))
```

### Другой метод разбиения

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

#Создаем объект-экземпляр класса
clf = KNeighborsClassifier(n_neighbors=34)

#Обучаем машину с помощью fit
clf.fit(X_train, y_train)

#Получаем прогноз с помощью predict
print("Прогнозы на тестовом наборе: ", clf.predict(X_test))

#Для оценки обобщающей способности модели мы вызываем метод score
print("Правильность на тестовом наборе: ", clf.score(X_test, y_test))
```
